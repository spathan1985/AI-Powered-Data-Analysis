```sql
with top10 as (
  select state_abbr, 
         sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double)) as total_emissions
  from emissions_data
  group by state_abbr
  order by total_emissions desc 
  limit 10
)
select sum(total_emissions),
  sum(total_emissions) / 
    (select sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double)) from emissions_data) * 100 as pct_of_total
from top10;
```
